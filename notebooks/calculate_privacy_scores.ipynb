```python
import pandas as pd

file_path1 = "user_study_2a_img.tsv"
file_path2 = "user_study_2b_img.tsv"

# Read the TSV file into a DataFrame
df1 = pd.read_csv(file_path1, sep='\t')
df2 = pd.read_csv(file_path2, sep='\t')

df = pd.concat([df1, df2])
df
```

```python
participant_cols = df.columns[2:]
df['average_score'] = df[participant_cols].mean(axis=1, skipna=True)

# Normalize to [0, 1]
df['normalized_score'] = (df['average_score'] - 1) / 4

# View result
df[['attribute_id', 'attribute_name', 'average_score', 'normalized_score']]
```

```python
df["attribute_id"].unique()
```

```python
target_attributes = [
    "a9_face_complete", "a74_address_current_complete", "a78_address_home_complete",
    "a103_license_plate_complete", "a109_person_body", "a13_full_nudity",
    "a21_name_last", "a18_ethnic_clothing", "a24_birth_date",
    "a26_handwriting", "a29_ausweis", "a30_credit_card", "a31_passport",
    "a32_drivers_license", "a33_student_id", "a35_mail", "a37_receipt",
    "a38_ticket", "a39_disability_physical", "a43_medicine", "a49_phone",
    "a70_education_history", "a73_landmark", "a7_fingerprint",
    "a82_date_time", "a85_username", "a8_signature", "a90_email"
]

# Filter the DataFrame
filtered_df = df[df['attribute_id'].isin(target_attributes)]

# Show result
filtered_df[['attribute_id', 'attribute_name', 'normalized_score']]
```

```python
# Get original min and max
min_score = filtered_df['normalized_score'].min()
max_score = filtered_df['normalized_score'].max()

# Scale to range [0.1, 1]
filtered_df['scaled_score'] = 0.1 + (filtered_df['normalized_score'] - min_score) / (max_score - min_score) * (1 - 0.1)
filtered_df['scaled_score'] = filtered_df['scaled_score'].round(2)

# View result
filtered_df[['attribute_name', 'normalized_score', 'scaled_score']]
```

```python
attribute_score_dict = dict(zip(filtered_df['attribute_id'], filtered_df['scaled_score']))
print(attribute_score_dict)
```

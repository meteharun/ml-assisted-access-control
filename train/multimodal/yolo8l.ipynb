```python
data_yaml = """
train: /kaggle/input/multimodal/images/train
val: /kaggle/input/multimodal/images/val
test: /kaggle/input/multimodal/images/test

nc: 7
names: ["credit_card", "passport", "drivers_license", "student_id", "mail", "receipt", "ticket"]
"""
with open("data.yaml", "w") as f:
    f.write(data_yaml.strip())
```

```python
!yolo detect train \
  model=yolov8l.pt \
  data=data.yaml \
  epochs=100 \
  imgsz=640 \
  batch=16 \
  workers=2 \
  project=yolo_multimodal \
  name=yolov8l_run1 \
  exist_ok=True \
  verbose=True \
  plots=True \
  save_json=True \
  save_period=5
```

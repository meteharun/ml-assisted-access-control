```python
fixed_yaml = """
path: /kaggle/input/d/meteharunakcay/visual-redactions
train: images/train
val: images/val
test: images/test

names:
  0: a105_face_all
  1: a108_license_plate_all
  2: a109_person_body
  3: a110_nudity_all
  4: a26_handwriting
  5: a39_disability_physical
  6: a43_medicine
  7: a7_fingerprint
  8: a8_signature
"""

with open("/kaggle/working/data.yaml", "w") as f:
    f.write(fixed_yaml)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8s-seg.pt")
model.train(
    data="/kaggle/working/data.yaml",
    epochs=50,
    imgsz=640,
    batch=8,
    name="kaggle_seg_run4",
    save_period=5
)
```

```python
from ultralytics import YOLO

# Load the trained model
model = YOLO("/kaggle/input/yolo-model/best.pt")

# Evaluate on validation set
metrics = model.val(data="/kaggle/working/data.yaml", save_json=True)
```
